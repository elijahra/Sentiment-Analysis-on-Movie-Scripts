```python
import string
import pandas as pd
import nltk as nl
from nltk import parse
import re
```

```python
file  = open("The Stranger.txt",'r')#saving the webscraped script to a text
the_stranger = file.read()
```

```python
sentences = re.split('\.', the_stranger)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer
```

```python
analyser = SentimentIntensityAnalyzer()
scores = []

for sentence in sentences:
    score = analyser.polarity_scores(sentence)
    scores.append(score)
```

```python
dataFrame = pd.DataFrame(scores)
```

```python
dataFrame.head(10)
```

```python
neg = dataFrame['neg'].rolling(200).mean()
pos = dataFrame['pos'].rolling(200).mean()
neu = dataFrame['neu'].rolling(200).mean()
comp = dataFrame['compound'].rolling(200).mean()
```

```python
plt.figure(figsize=(15, 10))

plt.plot(range(0,len(sentences)),neg)
plt.plot(range(0,len(sentences)),pos)
plt.plot(range(0,len(sentences)),neu)
plt.plot(range(0,len(sentences)),comp)
plt.ylim(-1,1)
plt.legend(prop={'size': 10});
```
